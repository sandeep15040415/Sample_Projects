## Data Preperation
1. Dropping irrelevant columns and rows
2. identifying the duplicate columns
3. renaming the columns
4. Feature creation.

```python
#removing unwanted columns
df_subset = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type',
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
      #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
    #'Opened', 
    #'Replaced by', 'Website',
     #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df_subset.columns
```

```python
df_subset.shape
```

```python
df_subset.dtypes
```

```python
### converting the opening_date_clean to date, as it is stored as object dtype
df_subset["opening_date_clean"] = pd.to_datetime(df_subset["opening_date_clean"])
```

```python
### checking if the data type has been changed for "opening_date_clean"
df_subset.dtypes
```

```python
### Renaming the columns by passing the dictionary 
df_subset.columns
df_subset = df_subset.rename(columns= {"coaster_name":"Coaster_Name",
                          "year_introduced" : "Year_Introduced",
                          "latitude":"Latitude",
                          "longitude":"Longitude",
                          "opening_date_clean" : "Opening_Date",
                          "speed_mph" : "Speed_mph",
                          "height_ft": "Height_ft",
                          "Inversions_clean": "Inversion",
                          "Gforce_clean" : "GForce"}
                )
```

```python
df_subset.head(5)
```

```python
### identify the missing values
df_subset.isna().sum()
```

```python
'''idenfying duplicate values using duplicated, duplicated will always ignore the 1st data and give 2nd row if its duplicate''' 
df_subset.loc[df_subset.duplicated()]
### identfied that we dont have duplicate records.
```

```python
### checking duplicate values based on columns.
df_subset.loc[df_subset.duplicated(subset="Coaster_Name")]
## we can see that we have duplicate values in column Coaster_Name, however need to further investigate for why we have 
# duplicates in Coaster_Name
```

```python
### investigating on Coaster_Name
df_subset[df_subset['Coaster_Name']=="Crystal Beach Cyclone"]
```

```python
'''from the above we can see that Crystal Beach Cyclone has 2 entries,most of the values in all the columns are 
identical, however only 1 column has different value i.e Year_Introduced''' 
### removing duplicates based on certain set of columns, if values are same in these columns, remove them
df_subset.columns
df_subset.duplicated(subset=["Coaster_Name","Location","Opening_Date"]).sum()
```

```python
## based on above onservation we will now take only those values which are not duplicated in 3 columns mentioned above
df_subset = df_subset.loc[~df_subset.duplicated(subset=["Coaster_Name","Location","Opening_Date"])] \
    .reset_index(drop = True)
### since we are dropping the rows, we need to make sure we are reseting the indexes.
```

```python
### checking if we have removed the duplicate rows using same example used above.
df_subset[df_subset["Coaster_Name"]== "Crystal Beach Cyclone"]
```

## Feature Understanding
##### Plotting Feature Distribution
        1. Histogram
        2. KDE
        3. Box plot

```python
### value_counts() will give us the distinct count of rollercoaster introduced for each year.
df_subset['Year_Introduced'].value_counts()
```

```python
### setting ab as axis so we can give the x axis name
ab = df_subset['Year_Introduced'].value_counts() \
.head(10) \
.plot(kind='bar',title = 'Top 10 years where Roller Coasters were Introduced')
ab.set_xlabel('Year Introduced')
ab.set_ylabel("Count of Roller Coasters")
plt.show()
```

```python
### destribution of speed in mph
ac = df_subset['Speed_mph'].plot(kind = 'hist',
                           bins = 20,
                           title = "Coaster Speed distribution (mph)")
ac.set_xlabel('Speed (mph)')
plt.show()
```

1. Note :From the above histogram,we can observer that the common speed of the coasters are between 40-60 mph
and there are some very rare speeds which is in the range on 140 mph

```python
### creating speed distrubution on kernel density plot
ak = df_subset['Speed_mph'].plot(kind='kde',
                           title = 'Coaster Speed distribution (mph)')
ak.set_xlabel('Speed (mph)')
plt.show()
```

### Feature relationships
        1. Scatter Plot
        2. Heatmap Correlation
        3. Pairplot
        4. Groupby Comparision

```python
ap = df_subset.plot(kind = 'scatter',
              x = 'Speed_mph',
              y = 'Height_ft',
              title = 'Roller Coaster Speed (mph) vs Height (ft)')
ap.set_xlabel('Speed (mph)')
ap.set_ylabel('Height (ft)')
plt.show()
```

```python
### similar plot, we will using seaborn
ax = sns.scatterplot(data=df_subset,
               x = "Speed_mph",
               y = "Height_ft",
               hue= "Type_Main"
                    )
ax.set_xlabel("Speed (mph)")
ax.set_ylabel("Height (ft)")
plt.show()
```

```python
### comparing features against each other using correlation
df_corr = df_subset[["Year_Introduced","Speed_mph","Height_ft","Inversion","GForce"]].dropna().corr()
```

```python
sns.heatmap(df_corr,annot= True)
plt.show()
```

### Ask a question about the data
- Trying to answer questions related to data
1. What are the locations with the fastest roller coasters (minimum of 10 coaster)?

```python
df_subset["Location"].value_counts()
##we can see below that location others is available with highest value, to further analysis we will remove location = other
```

```python
ax = df_subset.query('Location != "Other"') \
.groupby("Location")["Speed_mph"]\
.agg(['mean','count'])\
.query('count>= 10') \
.sort_values('mean')['mean'] \
.plot(kind = 'barh',figsize = (12,5),title = "Average Roller Coaster speed by location")
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
