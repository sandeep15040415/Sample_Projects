# Logistic Model for Cancer Patients
- In this project, data used is from Kaggle.
- The data set has lot of cells measurement which some doctors have made. Based on these measurements we need to check if the cells are M = malignant or B = benign

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
pd.set_option('display.max_columns', None)
```

```python
df_main = pd.read_csv("Patients_Data_set.csv")
df_main.head()
```

```python
df_main.shape
```

```python
df_main.dtypes
```

```python
df_main.isnull().sum()
```

```python
## Dropping unwanted column
df_main.drop(columns=["Unnamed: 32"],inplace=True)
df_main.columns
```

```python
## made a new data set, s
df_subset = df_main[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
## converting column diagnosis to category
df_subset["diagnosis"] = [1 if val =='M' else 0 for val in df_subset["diagnosis"] ]
df_subset["diagnosis"] = df_subset["diagnosis"].astype("category",copy = False)
df_subset.dtypes
```

- Based on the intial observation, the data is pretty clean, now begin with model preperation.
- Dividing the data in predictors variable Target variables.
- Then divide the Data into Test and Train.

```python
y = df_subset["diagnosis"]
X = df_subset.drop(columns="diagnosis")
# for X variable, i am taking all the variables apart from diagnosis.
```

# Normalizing Data
- Normalizing the data in X data set, as the X data set has lot of measure values whose units are not uniform when compared to other variables in X. For model to have unformity in the measures, we need to Normalize the values in X .

```python
from sklearn.preprocessing import StandardScaler
## creating a scalar object 
scalar = StandardScaler()
# fit the scalar to the data in x
X_norm = scalar.fit_transform(X)
X_norm
```

## Splitting the data to Test and Train

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X_norm,y,test_size= 0.30,random_state=42)
```

## Train the model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)

## predict the target variable based on the test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

## Model evaluation

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy:,{accuracy: .2f}")
```

## Exporting the data along with the original data frame

```python
df_main["Predicted"] = lr.predict(X_norm)
df_main["Predicted_Diagnosis"] = ["M" if p == 1 else "B" for p in df_main["Predicted"]]
df_main.head(2)
```

```python
df_main.to_csv('Cancer_Patients_Diagnosis_Prediction.csv', index=False)
```
